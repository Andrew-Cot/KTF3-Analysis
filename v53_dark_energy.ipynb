```python
import numpy as np
import matplotlib.pyplot as plt

# === FUNDAMENTAL CONSTANTS ===
c      = 2.998e8          # m/s
hbar   = 1.055e-34        # J·s
G      = 6.674e-11        # m³/(kg·s²)
Mpc    = 3.086e22         # m per Mpc

# === KTF³ PARAMETERS ===
H0_kms = 67.1             # km/s/Mpc
H0     = H0_kms * 1e3 / Mpc   # s⁻¹
T1_Mpc = 1660.0           # Mpc
T2_Mpc = 2630.0
T3_Mpc = 2750.0
T23_Mpc = (T2_Mpc + T3_Mpc) / 2
T1     = T1_Mpc * Mpc     # m
T23    = T23_Mpc * Mpc    # m
theta_R = 25.7            # degrees
N_cycles = 360.0 / theta_R  # = 14.007
R_H    = c / H0           # Hubble radius in m
Omega_Lambda = 0.685

# === OBSERVED Λ ===
Lambda_obs = 3 * H0**2 * Omega_Lambda / c**2

print('=' * 60)
print('v53 — Dark Energy from KTF³ Topology')
print('=' * 60)
print(f'H₀          = {H0_kms} km/s/Mpc = {H0:.4e} s⁻¹')
print(f'T₁          = {T1_Mpc} Mpc = {T1:.4e} m')
print(f'T₂₃         = {T23_Mpc} Mpc')
print(f'R_H = c/H₀  = {R_H/Mpc:.0f} Mpc = {R_H:.4e} m')
print(f'θ_R         = {theta_R}°')
print(f'N_cycles    = {N_cycles:.3f}')
print(f'Ω_Λ         = {Omega_Lambda}')
print()
print(f'Λ_obs       = {Lambda_obs:.4e} m⁻²')
print(f'  = 3H₀²Ω_Λ/c²')
```

```python
# === TEST ALL CANDIDATE FORMULAS ===

results = []

# Formula 1: Quantum zero-point cutoff at T1
# Vacuum energy integral cut off at wavelength T1
L1 = hbar * c / T1**4
ratio1 = L1 / Lambda_obs
results.append(('ℏc / T₁⁴', L1, ratio1, 'Quantum cutoff at T₁'))

# Formula 2: Geometric — one cycle on 14-fold structure
# Each cycle covers angle 2π/14, projected on T1
L2 = (2*np.pi / N_cycles)**2 / T1**2
ratio2 = L2 / Lambda_obs
results.append(('(2π/N)² / T₁²', L2, ratio2, 'Geometric: 14-cycle on T₁'))

# Formula 3: Pure Hubble scale
L3 = H0**2 / c**2
ratio3 = L3 / Lambda_obs
results.append(('H₀² / c²', L3, ratio3, 'Pure Hubble scale'))

# Formula 4: Hubble radius squared
L4 = 1.0 / R_H**2
ratio4 = L4 / Lambda_obs
results.append(('1 / R_H²', L4, ratio4, 'Hubble radius'))

# Formula 5: Mixed — Hubble × topology
# θ_R connects H0 and T1/T23 → use geometric mean
T_geom = np.sqrt(T1 * T23)
L5 = (H0/c)**2 * (T_geom / R_H)**2 / T1**2
ratio5 = L5 / Lambda_obs
results.append(('(H₀/c)²·(T_geom/R_H)²/T₁²', L5, ratio5, 'Mixed topological-expansion'))

# Formula 6: θ_R² / R_H²
# The precessional angle over the Hubble radius
L6 = np.radians(theta_R)**2 / R_H**2
ratio6 = L6 / Lambda_obs
results.append(('θ_R² / R_H²', L6, ratio6, 'Precessional angle / Hubble²'))

# Formula 7: 3Ω_Λ × H0²/c² (trivially exact — for reference)
L7 = 3 * Omega_Lambda * H0**2 / c**2
ratio7 = L7 / Lambda_obs
results.append(('3Ω_Λ H₀²/c² (trivial)', L7, ratio7, 'Definition (exact by construction)'))

# Formula 8: (H0·T1/c)² / T1² = H0²/c²  — same as F3 but written differently
# Try: arctan(H0·T1/c)² / T1² — using the D11v4 angle directly
alpha1 = np.arctan(H0 * T1 / c)   # = arctan(0.371) = 20.4°
alpha2 = np.arctan(H0 * T23 / c)  # = arctan(0.601) = 31.0°
theta_rad = np.radians(theta_R)
L8 = theta_rad**2 / T1**2
ratio8 = L8 / Lambda_obs
results.append(('θ_R² / T₁²', L8, ratio8, 'Precessional angle / T₁²'))

print('=' * 75)
print(f'{"Formula":<30} {"Λ [m⁻²]":>14} {"Ratio to obs":>14} {"log10(ratio)":>12}')
print('-' * 75)
for name, val, ratio, desc in results:
    log_ratio = np.log10(abs(ratio)) if ratio > 0 else float('nan')
    marker = ' ✓' if abs(np.log10(abs(ratio))) < 0.5 else ('  ~' if abs(np.log10(abs(ratio))) < 1.5 else '')
    print(f'{name:<30} {val:>14.4e} {ratio:>14.4f} {log_ratio:>12.2f}{marker}')
print('=' * 75)
print(f'Λ_obs = {Lambda_obs:.4e} m⁻²  (reference)')
print()
print('✓ = within factor 3 (log10 < 0.5)')
print('~ = within factor 30 (log10 < 1.5)')
```

```python
# === PARAMETER SCAN: which combination of T1, N_cycles gives exact Λ? ===
# If formula 2 is correct: Λ = (2π/N)² / T1²
# → N_eff = 2π / sqrt(Λ_obs * T1²)

N_eff = 2*np.pi / np.sqrt(Lambda_obs * T1**2)
print(f'For formula (2π/N)²/T₁² to give Λ_obs exactly:')
print(f'  N_eff = {N_eff:.4f}  (observed = {N_cycles:.3f})')
print(f'  Ratio N_eff/N_obs = {N_eff/N_cycles:.4f}')
print()

# What T1 would make formula 2 exact?
T1_exact_Mpc = (2*np.pi / N_cycles) / np.sqrt(Lambda_obs) / Mpc
print(f'For N=14.007 exact: T₁ needed = {T1_exact_Mpc:.0f} Mpc  (measured = {T1_Mpc} Mpc)')
print(f'  Difference: {abs(T1_exact_Mpc - T1_Mpc):.0f} Mpc = {abs(T1_exact_Mpc/T1_Mpc - 1)*100:.1f}%')
print()

# Best fit: scan over simple formulas Λ = α / T1^n / R_H^m
print('Scanning: Λ = (2π/N)^a / T₁^b / R_H^(2-b)  for integer a,b...')
best_ratio = 1e10
best_params = None
for a in [1, 2, 3, 4]:
    for b in [0, 1, 2, 3, 4]:
        m = 2 - b
        val = (2*np.pi/N_cycles)**a / T1**b / R_H**m if m >= 0 else (2*np.pi/N_cycles)**a / T1**b * R_H**(-m)
        ratio = val / Lambda_obs
        if 0 < ratio and abs(np.log10(ratio)) < abs(np.log10(best_ratio)):
            best_ratio = ratio
            best_params = (a, b, m, val)

a, b, m, val = best_params
print(f'Best fit: Λ = (2π/N)^{a} / T₁^{b} / R_H^{m}')
print(f'  Value = {val:.4e} m⁻²')
print(f'  Ratio = {val/Lambda_obs:.4f}')
print(f'  log10(ratio) = {np.log10(val/Lambda_obs):.3f}')
```
