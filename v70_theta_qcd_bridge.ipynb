# v70 — θ_R = π/b₀: The QCD-Cosmology Bridge
**Hypothesis:** The KTF³-R rotation angle θ_R is not empirical — it follows from QCD.
**Formula:** θ_R = 180°/b₀ where b₀ = 11 − 2N_f/3 (QCD β-function)
**With N_f=6:** b₀ = 7 → θ_R = 180°/7 = 25.714°
**Observed:** θ_R = 25.7° (from CMB ℓ*=3.5)
**Deviation:** 0.054% — well within measurement uncertainty
**OSF:** osf.io/42nks | GitHub: github.com/Andrew-Cot/KTF3-Analyse


```python
import numpy as np
import matplotlib.pyplot as plt

print('=== v70: THETA_R = PI/b0 QCD BRIDGE ===')
print()

# QCD beta function
# b0 = 11 - 2*Nf/3
# For Nf = 3,4,5,6 quarks
print('QCD beta function coefficient b0 = 11 - 2*Nf/3:')
print()
print('Nf   b0      theta_R=180/b0   N=360/theta_R   Match CMB?')
print('-'*65)

theta_CMB = 25.7  # observed from CMB
N_CMB = 360.0 / theta_CMB

for Nf in [3, 4, 5, 6]:
    b0 = 11 - 2*Nf/3
    theta_R = 180.0 / b0
    N = 360.0 / theta_R
    dev = abs(theta_R - theta_CMB) / theta_CMB * 100
    match = '<-- MATCH!' if dev < 1.0 else ''
    print(str(Nf) + '    ' + str(round(b0,4)) + '   ' + str(round(theta_R,4)) + 'deg        N=' + str(round(N,4)) + '      dev=' + str(round(dev,3)) + '%  ' + match)

print()
print('=== KEY RESULT ===')
b0 = 7.0  # Nf=6
theta_pred = 180.0 / b0
theta_obs  = 25.7
N_pred = 360.0 / theta_pred
N_obs  = 360.0 / theta_obs
dev_theta = abs(theta_pred - theta_obs) / theta_obs * 100
dev_N     = abs(N_pred - N_obs) / N_obs * 100

print('QCD prediction (Nf=6, b0=7):')
print('  theta_R = 180/7 = ' + str(round(theta_pred,6)) + ' deg')
print('  N = 360/theta_R = ' + str(round(N_pred,6)))
print()
print('CMB observation:')
print('  theta_R = ' + str(theta_obs) + ' deg  (from ell*=3.5)')
print('  N = ' + str(round(N_obs,6)))
print()
print('Deviation: theta_R = ' + str(round(dev_theta,4)) + '%')
print('Deviation: N       = ' + str(round(dev_N,4)) + '%')
print()
print('INTERPRETATION:')
print('  If theta_R = pi/b0, then N = 2*b0 = 14')
print('  This connects:')
print('    KTF3-R (cosmological CMB angle)')
print('    Twin Barrier (QCD beta function)')
print('    through one equation: theta_R = pi/b0')
```

```python
# Derive alpha from theta_R = pi/b0
print('=== DERIVING ALPHA FROM QCD ===')
print()
print('If theta_R = pi/b0, then:')
print('  N = 2*b0 = 14')
print('  alpha = N * 1.515 = 14 * 1.515 = 21.21  (Twin Barrier)')
print()
print('And from KTF3-R:')
print('  alpha = ln(1/eta_B)')
print()
print('If both are true, then:')
print('  ln(1/eta_B) = 2*b0 * 1.515')
print('  eta_B = e^{-2*b0*1.515}')
print()

b0 = 7.0
factor = 1.515
alpha_pred = 2 * b0 * factor
eta_B_pred = np.exp(-alpha_pred)
eta_B_obs  = 6e-10
dev_eta = abs(eta_B_pred - eta_B_obs) / eta_B_obs * 100

print('Predicted eta_B = e^{-' + str(round(alpha_pred,4)) + '} = ' + '{:.3e}'.format(eta_B_pred))
print('Observed  eta_B = 6.0e-10 (Planck 2018)')
print('Deviation: ' + str(round(dev_eta,1)) + '%')
print()

# What factor makes it exact?
factor_exact = np.log(1/eta_B_obs) / (2*b0)
print('For exact match, the AdS sector factor should be:')
print('  1.515_exact = ln(1/eta_B) / (2*b0) = ' + str(round(factor_exact,6)))
print('  1.515_TB    = 1.515000 (Radojicic 2026)')
print('  Deviation:    ' + str(round(abs(factor_exact-1.515)/1.515*100,4)) + '%')
print()
print('CONCLUSION:')
print('  theta_R = pi/b0 is the bridge between:')
print('    - QCD beta function (particle physics)')
print('    - CMB precessional angle (cosmology)')
print('  If the AdS sector factor 1.515 can be derived')
print('  from topology, eta_B is predicted from first principles.')
```

```python
# Plot
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('v70 -- theta_R = pi/b0: QCD-Cosmology Bridge', fontweight='bold')

# Left: theta_R vs Nf
ax = axes[0]
Nf_vals = np.array([3, 4, 5, 6])
b0_vals = 11 - 2*Nf_vals/3
theta_vals = 180.0 / b0_vals
N_vals = 360.0 / theta_vals

ax.plot(Nf_vals, theta_vals, 'bo-', lw=2, markersize=10, label='theta_R = 180/b0')
ax.axhline(25.7, color='red', ls='--', lw=2, label='CMB observed: 25.7 deg')
ax.axhline(180/7, color='green', ls=':', lw=2, label='Nf=6: 180/7 = 25.714 deg')
ax.scatter([6], [180/7], c='green', s=200, zorder=5)
ax.set_xlabel('Number of quark flavours Nf')
ax.set_ylabel('Predicted theta_R [deg]')
ax.set_title('QCD prediction vs CMB observation')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)

# Right: N = 2*b0
ax = axes[1]
N_pred_vals = 2 * b0_vals
ax.plot(Nf_vals, N_pred_vals, 'bo-', lw=2, markersize=10, label='N = 2*b0')
ax.axhline(360/25.7, color='red', ls='--', lw=2, label='CMB observed: N=14.007')
ax.axhline(14, color='green', ls=':', lw=2, label='Nf=6: N=14 exactly')
ax.scatter([6], [14], c='green', s=200, zorder=5)
ax.set_xlabel('Number of quark flavours Nf')
ax.set_ylabel('Predicted N = 360/theta_R')
ax.set_title('N = 2*b0: closes to 14 for Nf=6')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('v70_theta_qcd_bridge.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: v70_theta_qcd_bridge.png')
```
