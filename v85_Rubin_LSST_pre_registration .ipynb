## 2. Prediction 1 — Galaxy Spin Alignment

**Claim.** LSST DR1 spiral-galaxy angular-momentum vectors will show statistically significant alignment with the preferred axis **(l=277°, b=-31°)** in the galactic frame.

**Prior motivation.** GZ2 winding analysis (v80) on SDSS spirals (N≈106k) showed σ=+1.07 consistent with this axis. LSST DR1 provides roughly 50× larger sample and deeper redshift coverage.

**Test statistic.** Dipole amplitude D of the spin-vector distribution projected onto the unit sphere, compared against isotropic Monte Carlo null.

```python
# --- Prediction 1 thresholds ---
P1_sigma_threshold = 3.0         # confirmation threshold
P1_axis_tolerance_deg = 10.0     # aligned axis must be within 10° of (277°, -31°)
P1_min_sample = 1e8              # minimum spiral sample size for valid test

print("PREDICTION 1 — Spin Alignment")
print("-" * 50)
print(f"Claim:          Dipole σ ≥ {P1_sigma_threshold}")
print(f"Axis constraint: within {P1_axis_tolerance_deg}° of (l=277°, b=-31°)")
print(f"Sample floor:   N ≥ {P1_min_sample:.0e} spiral classifications")
print(f"\nFalsifier:      σ < 1.5  OR  axis >20° away")
print(f"Confirmation:   σ ≥ 3.0  AND axis within tolerance")
```

## 3. Prediction 2 — Chiral (Helicity) Asymmetry

**Claim.** The fraction of clockwise vs. counter-clockwise spirals, projected along the preferred axis, deviates from 50/50 by **A > 0.02** (i.e., ≥2% asymmetry).

**Prior motivation.** The ℤ₁₄ screw topology is intrinsically chiral — the rotation θ_R = π/b₀ selects a handedness. Shamir (2020, 2021) reported approximately 1% asymmetries in SDSS that were debated in the literature; LSST's uniform coverage and automated morphology will clarify.

**Test statistic.** A = (N_CW − N_CCW) / (N_CW + N_CCW), measured in the preferred-axis hemisphere.

```python
# --- Prediction 2 thresholds ---
P2_A_threshold = 0.02            # minimum asymmetry
P2_sigma_threshold = 3.0         # minimum detection significance
P2_sign_prediction = +1          # +1 = CW excess along +axis direction
                                 # (sign fixed by ℤ₁₄ convention, see D0 v2)

print("PREDICTION 2 — Chiral Asymmetry")
print("-" * 50)
print(f"Claim:          |A| ≥ {P2_A_threshold:.3f}  at σ ≥ {P2_sigma_threshold}")
print(f"Sign:           A > 0 (CW excess in +axis hemisphere)")
print(f"\nFalsifier:      A consistent with zero OR opposite sign at σ>2")
print(f"Confirmation:   A ≥ +0.02 at σ ≥ 3")
```

## 4. Prediction 3 — Redshift Amplification of Asymmetry

**Claim.** The chiral asymmetry **grows** with redshift: dA/dz > 0 over 0.1 < z < 1.0.

**Prior motivation.** In the ℤ₁₄ framework, the rotational identification φ_R(x,y,z,t) = R(θ_R)·(x,y,z), t+T₁ implies that the topological signature accumulates over comoving distance. Deeper redshift probes longer path lengths through the screw structure, so the asymmetry should amplify. LSST DR1 will reach z≈1.2 for bright spirals — a regime inaccessible to SDSS.

**Test statistic.** Linear slope of A(z) across four redshift bins (0.1–0.3, 0.3–0.5, 0.5–0.8, 0.8–1.2).

```python
# --- Prediction 3 thresholds ---
P3_bins = [(0.1, 0.3), (0.3, 0.5), (0.5, 0.8), (0.8, 1.2)]
P3_min_per_bin = 5e6             # minimum galaxies per bin
P3_slope_threshold = 0.01        # minimum dA/dz [per unit z]
P3_sigma_threshold = 2.5         # minimum slope significance

print("PREDICTION 3 — Redshift Amplification")
print("-" * 50)
print(f"Claim:          dA/dz ≥ {P3_slope_threshold}  at σ ≥ {P3_sigma_threshold}")
print(f"Redshift bins:  {P3_bins}")
print(f"Sample floor:   ≥ {P3_min_per_bin:.0e} galaxies per bin")
print(f"\nFalsifier:      dA/dz ≤ 0 at σ>2  OR  non-monotonic pattern")
print(f"Confirmation:   Monotonic increase, slope ≥ +0.01 at σ ≥ 2.5")
```

## 5. Prediction 4 — Spatial Correlation Peak at T₁ = 1660 Mpc

**Claim.** The spiral-galaxy two-point spin correlation function ξ_spin(r) will show an excess **peak** at the fundamental period r ≈ T₁ = 1660 Mpc, and smaller peaks at the harmonic scales r ≈ 415 Mpc (n=4 mode) and r ≈ 830 Mpc (n=2 mode).

**Prior motivation.** The wave equation on ℤ₁₄ permits only sine modes. Predicted scales: 166, 277, 415, 830, 1315, 1660 Mpc. The Sloan Great Wall (≈420 Mpc) already matches the n=4 mode at 1.2% deviation. LSST DR1 is the first dataset with enough depth × area to resolve the fundamental 1660 Mpc peak cleanly.

**Test statistic.** Wavelet decomposition of ξ_spin(r), peak-finding against smooth ΛCDM background.

```python
# --- Prediction 4: expected peak scales ---
predicted_scales_Mpc = {
    "n=10 (T₁/10)": 166,
    "n=6  (T₁/6)":  277,
    "n=4  (T₁/4)":  415,
    "n=2  (T₁/2)":  830,
    "n=4/3 (T₁·3/4)": 1315,
    "n=1  (T₁)":    1660,
}

P4_tolerance_pct = 3.0           # ±3% tolerance on peak position
P4_sigma_threshold = 2.0         # per-peak detection
P4_required_peaks = 3            # at least 3 of 6 must confirm

print("PREDICTION 4 — Spatial Correlation Peaks")
print("-" * 50)
print(f"{'Mode':<20} {'Expected [Mpc]':>15}")
for mode, scale in predicted_scales_Mpc.items():
    print(f"{mode:<20} {scale:>15}")
print()
print(f"Tolerance:      ±{P4_tolerance_pct}% on peak position")
print(f"Per-peak σ:     ≥ {P4_sigma_threshold}")
print(f"Required:       ≥ {P4_required_peaks} of 6 peaks confirmed")
print(f"\nFalsifier:      Fewer than 2 peaks confirmed")
print(f"Confirmation:   ≥ 3 peaks including n=1 (T₁) or n=2 (T₁/2)")
```

## 6. Combined Decision Rule

To avoid cherry-picking, a **combined decision rule** is pre-registered:

| Outcome          | Criterion                             | Interpretation                            |
| ---------------- | ------------------------------------- | ----------------------------------------- |
| **Strong support** | ≥ 3 of 4 predictions confirmed       | KTF³-R favored over ΛCDM isotropy         |
| **Mixed**          | Exactly 2 of 4 confirmed             | Partial support, theoretical review       |
| **Weak support**   | 1 of 4 confirmed                     | Near-null, likely coincidence             |
| **Falsified**      | 0 of 4 confirmed, ≥ 2 opposite-sign at σ>2 | KTF³-R as formulated is wrong       |

A prediction is "confirmed" if its listed σ-threshold is met **and** the sign matches the pre-registered claim.

```python
def evaluate_LSST_outcome(P1_sigma, P2_A, P2_sigma, P3_slope, P3_sigma,
                          P4_peaks_confirmed, P4_has_T1_or_T1half):
    """
    Pre-registered decision function.
    Call this ONCE with LSST DR1 results; do not tune.
    """
    conf = 0
    opposite = 0

    # P1
    if P1_sigma >= P1_sigma_threshold: conf += 1
    elif P1_sigma <= -2.0: opposite += 1

    # P2
    if P2_A >= P2_A_threshold and P2_sigma >= P2_sigma_threshold: conf += 1
    elif P2_A <= -P2_A_threshold and P2_sigma >= 2.0: opposite += 1

    # P3
    if P3_slope >= P3_slope_threshold and P3_sigma >= P3_sigma_threshold: conf += 1
    elif P3_slope <= -P3_slope_threshold and P3_sigma >= 2.0: opposite += 1

    # P4
    if P4_peaks_confirmed >= P4_required_peaks and P4_has_T1_or_T1half: conf += 1
    elif P4_peaks_confirmed < 2: opposite += 0  # P4 can't give opposite sign

    # Verdict
    if conf >= 3:
        return "STRONG SUPPORT", conf, opposite
    if conf == 2:
        return "MIXED — theoretical review", conf, opposite
    if conf == 1:
        return "WEAK SUPPORT — likely coincidence", conf, opposite
    if conf == 0 and opposite >= 2:
        return "FALSIFIED", conf, opposite
    return "NULL — insufficient signal", conf, opposite


# --- Example (mock) call: DO NOT run on real LSST data until DR1 release ---
print("Example call with hypothetical values:")
print(evaluate_LSST_outcome(
    P1_sigma=3.2, P2_A=0.025, P2_sigma=3.5,
    P3_slope=0.015, P3_sigma=2.8,
    P4_peaks_confirmed=4, P4_has_T1_or_T1half=True
))
```
