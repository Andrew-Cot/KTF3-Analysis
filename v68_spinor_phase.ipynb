# v68 — Spinor Phase after 14 Cycles = −1
**Question:** After 14 rotations of 25.7°, does a spinor pick up phase −1?
**Theory:** Spinor phase = e^{i·k·θ_R/2} after k rotations
**After k=14:** e^{i·14·θ_R/2} = e^{i·14·12.85°} = e^{i·179.9°} ≈ e^{iπ} = −1
**This is the origin of fermionic boundary conditions in KTF³-R**
**OSF:** osf.io/42nks | GitHub: github.com/Andrew-Cot/KTF3-Analyse

```python
import numpy as np
import matplotlib.pyplot as plt

theta_R = 25.7  # degrees per cycle
N = 14          # cycles

print('=== SPINOR PHASE AFTER k CYCLES ===')
print()
print('For a rotation by angle θ, a spinor picks up phase e^{iθ/2}')
print('(Spinors are double-valued: 4π rotation = identity, 2π = -1)')
print()
print(f'KTF³-R: θ_R = {theta_R}° per cycle')
print(f'Spinor phase per cycle: e^{{iθ_R/2}} = e^{{i·{theta_R/2}°}}')
print()
print(f'{"k":>3}  {"Total angle":>13}  {"Spinor phase (half angle)":>26}  {"Re(phase)":>11}  {"Im(phase)":>11}  {"≈ ?":<10}')
print('-'*80)

for k in range(1, 15):
    total_angle_deg = k * theta_R
    half_angle_rad  = np.radians(k * theta_R / 2)
    phase = np.exp(1j * half_angle_rad)
    
    approx = ''
    if abs(phase - 1)   < 0.01: approx = '≈ +1'
    if abs(phase - (-1))< 0.05: approx = '≈ −1 ← FERMIONIC!'
    if abs(phase - 1j)  < 0.05: approx = '≈ +i'
    if abs(phase + 1j)  < 0.05: approx = '≈ −i'
    
    marker = ' <<<' if k == 14 else ''
    print(f'{k:>3}  {total_angle_deg:>10.2f}°   e^{{i·{k*theta_R/2:.2f}°}}  {phase.real:>11.6f}  {phase.imag:>11.6f}  {approx:<10}{marker}')

print()
phase_14 = np.exp(1j * np.radians(14 * theta_R / 2))
print(f'After k=14 cycles:')
print(f'  Total rotation: {14*theta_R:.2f}°')
print(f'  Half angle: {14*theta_R/2:.2f}°')
print(f'  Spinor phase: e^{{i·{14*theta_R/2:.2f}°}} = {phase_14.real:.6f} + {phase_14.imag:.6f}i')
print(f'  |phase - (-1)| = {abs(phase_14 - (-1)):.6f}')
print()
print(f'RESULT: After 14 cycles, spinor phase ≈ −1  ✅')
print(f'  This means: ψ(x, t+14T₁) = −ψ(x, t)')
print(f'  Fermionic boundary condition! (anti-periodic in 14 cycles)')
```

```python
# Visualise spinor phase on unit circle
thetas_k = [np.radians(k * theta_R / 2) for k in range(15)]
phases_k  = [np.exp(1j*t) for t in thetas_k]

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('v68 — Spinor Phase After k Cycles (KTF³-R)', fontweight='bold')

# Unit circle plot
ax = axes[0]
circle = plt.Circle((0,0), 1, fill=False, color='gray', lw=1.5, ls='--')
ax.add_patch(circle)
ax.axhline(0, color='gray', lw=0.5)
ax.axvline(0, color='gray', lw=0.5)

for k, ph in enumerate(phases_k):
    if k == 0:
        continue
    color = 'green' if k == 14 else ('red' if k == 7 else 'steelblue')
    size  = 150 if k == 14 else (100 if k == 7 else 50)
    ax.scatter(ph.real, ph.imag, c=color, s=size, zorder=5)
    if k in [1, 7, 14]:
        ax.annotate(f'k={k}', (ph.real+0.05, ph.imag+0.05), fontsize=9,
                   color=color, fontweight='bold')

ax.scatter(-1, 0, c='red', s=200, marker='*', zorder=6, label='−1 (target)')
ax.set_xlim(-1.3, 1.3)
ax.set_ylim(-1.3, 1.3)
ax.set_aspect('equal')
ax.set_xlabel('Re(phase)')
ax.set_ylabel('Im(phase)')
ax.set_title('Spinor phase e^{ikθ_R/2} on unit circle')
ax.legend(fontsize=9)
ax.grid(alpha=0.2)

# Phase vs k
ax = axes[1]
k_vals = range(1, 15)
re_phases = [np.exp(1j*np.radians(k*theta_R/2)).real for k in k_vals]
ax.plot(k_vals, re_phases, 'bo-', lw=2, markersize=8)
ax.axhline(-1, color='red', ls='--', lw=2, label='−1 (fermionic)')
ax.axhline(+1, color='green', ls='--', lw=2, label='+1 (bosonic)')
ax.axhline(0, color='gray', ls=':', lw=1)
ax.axvline(14, color='purple', ls=':', lw=2, label='k=14')
ax.scatter([14], [re_phases[13]], c='purple', s=200, zorder=5)
ax.set_xlabel('Number of cycles k')
ax.set_ylabel('Re(spinor phase)')
ax.set_title(f'Re(e^{{ikθ_R/2}}) — reaches −1 at k≈14')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('v68_spinor_phase.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: v68_spinor_phase.png')
```
