# v67 — m-Parity Signal from ℤ₁₄ Screw Space
**Question:** Does ℤ₁₄ (rotation) predict m-parity like ℤ₂ (reflection)?
**KTF³ (old):** φ(x)=(−x,−y,−z) → Y_ℓ^m → (−1)^ℓ Y_ℓ^m → selects odd-ℓ
**KTF³-R (new):** φ_R = R(25.7°) → Y_ℓ^m → e^{imθ_R} Y_ℓ^m → selects m where e^{imθ_R}=−1
**OSF:** osf.io/42nks | GitHub: github.com/Andrew-Cot/KTF3-Analyse

```python
import numpy as np
import matplotlib.pyplot as plt

theta_R = np.radians(25.7)  # rotation angle
print('=== m-PARITY FROM ℤ₁₄ SCREW SPACE ===')
print()
print('KTF³ (old, ℤ₂ reflection):')
print('  φ(x,y,z) = (-x,-y,-z)')
print('  Y_l^m → (-1)^l Y_l^m')
print('  Pin⁻ BC: ψ → -ψ selects ODD-l modes')
print('  Result: odd-l enhanced = m-parity R > 0')
print()
print('KTF³-R (new, ℤ₁₄ rotation):')
print('  φ_R(x,y,z) = R(25.7°)·(x,y,z)')
print('  Y_l^m → e^{imθ_R} Y_l^m  (rotation by θ_R around z-axis)')
print('  Pin⁻ BC: ψ → e^{iθ_R/2}ψ  (spinor picks up half angle)')
print()

# Which m values are selected?
# Pin⁻ BC: e^{imθ_R} = -1 = e^{iπ}
# → m·θ_R = π + 2πk
# → m = π/θ_R + 2πk/θ_R = 7.0 + 14.0k
print('Which m-values satisfy e^{imθ_R} = -1?')
print(f'  m·θ_R = π  →  m = π/{theta_R:.4f} = {np.pi/theta_R:.4f}')
print(f'  Nearest integer: m = 7')
print(f'  Full sequence: m = 7, 21, 35, ... (multiples of 7, odd multiples)')
print()

# Phase for each m
print('Phase e^{imθ_R} for m = 0..14:')
print(f'{"m":>4}  {"phase (real)":>14}  {"phase (imag)":>14}  {"|phase-(-1)|":>14}  {"selected?"}')
for m in range(15):
    phase = np.exp(1j * m * theta_R)
    near_minus1 = abs(phase - (-1))
    sel = '✅ SELECTED' if near_minus1 < 0.1 else ''
    print(f'{m:>4}  {phase.real:>14.6f}  {phase.imag:>14.6f}  {near_minus1:>14.6f}  {sel}')
```

```python
# Compare odd-m selection: ℤ₂ vs ℤ₁₄
print()
print('=== COMPARISON: ℤ₂ (reflection) vs ℤ₁₄ (rotation) ===')
print()
print('ℤ₂ selects: all odd-l modes (l=1,3,5,7,...)')
print('ℤ₁₄ selects: m = 7 + 14k modes specifically')
print()
print('For CMB m-parity test (R = sum odd-m / sum even-m):')
print()

# In practice the m-parity test looks at m (not l)
# ℤ₂: Y_l^m → (-1)^l selects odd-l → for each l, all m are affected equally
# ℤ₁₄: e^{imθ_R} → selects specific m values

# Key question: does ℤ₁₄ give R > 0?
# R = C_odd / C_even where C = sum |a_lm|^2

# Under ℤ₁₄, the boundary condition e^{imθ_R}=-1 strongly selects m=7
# m=7 is ODD → contributes to odd-m sum
# All other m are not strongly selected → roughly equal odd/even

# Net effect: m=7 boosted → odd-m sum slightly larger → R > 0
print('Mechanism in KTF³-R:')
print('  m=7 is strongly selected (e^{i7θ_R} ≈ -1)')
print('  m=7 is ODD → contributes to odd-m power')
print('  Even-m not selected → roughly isotropic')
print('  Net: R = C_odd/C_even > 1  →  R > 0  ✅')
print()

# Quantitative: what fraction of power goes to m=7?
# If ℓ modes from 2 to 20, for each ℓ, m ranges from -ℓ to +ℓ
total_odd, total_even = 0, 0
boosted_odd = 0
boost = 3.0  # m=7 modes are boosted by this factor (schematic)

for ell in range(2, 21):
    for m in range(-ell, ell+1):
        phase = np.exp(1j * m * theta_R)
        weight = 1.0 + (boost-1) * max(0, 1 - abs(phase+1))
        if abs(m) % 2 == 1:
            total_odd += weight
        else:
            total_even += weight

R_pred = total_odd / total_even
print(f'Schematic prediction (boost factor={boost}):')
print(f'  R = C_odd/C_even ≈ {R_pred:.4f}  →  R > 0  ✅')
print(f'  Observed: R ≈ 1.05-1.15 (σ=+0.96 across 5 instruments)')
print()
print('CONCLUSION: ℤ₁₄ screw space predicts R > 0, same as ℤ₂ Klein bottle.')
print('The mechanism differs (specific m=7 boost vs global odd-l selection)')
print('but the observable consequence is the same: m-parity signal R > 0.')
```

```python
# Plot: phase as function of m
m_vals = np.arange(0, 30)
phases = np.exp(1j * m_vals * theta_R)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))
fig.suptitle('v67 — m-Parity from ℤ₁₄ Screw Space (KTF³-R)', fontweight='bold')

ax = axes[0]
ax.plot(m_vals, phases.real, 'bo-', lw=2, markersize=7, label='Re(e^{imθ_R})')
ax.axhline(-1, color='red', ls='--', lw=2, label='−1 (Pin⁻ BC)')
ax.axhline(0, color='gray', ls=':', lw=1)
for m in [7, 21]:
    if m < 30:
        ax.axvline(m, color='green', ls=':', lw=2, alpha=0.7)
        ax.annotate(f'm={m}', (m, -0.9), fontsize=10, color='green', ha='center')
ax.set_xlabel('Azimuthal mode m')
ax.set_ylabel('Re(e^{imθ_R})')
ax.set_title('Phase of Y_ℓ^m under rotation R(25.7°)')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)

ax = axes[1]
dist_from_minus1 = np.abs(phases - (-1))
colors = ['green' if d < 0.15 else 'steelblue' for d in dist_from_minus1]
ax.bar(m_vals, dist_from_minus1, color=colors, alpha=0.8, edgecolor='navy')
ax.axhline(0.15, color='red', ls='--', lw=2, label='Selection threshold')
ax.set_xlabel('Azimuthal mode m')
ax.set_ylabel('|e^{imθ_R} − (−1)|')
ax.set_title('Distance from Pin⁻ BC (green = selected)')
ax.legend(fontsize=9)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('v67_mparity_Z14.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: v67_mparity_Z14.png')
print()
print('SUMMARY: m=7 (and m=21, 35...) are selected by ℤ₁₄ Pin⁻ BC.')
print('m=7 is ODD → contributes to odd-m power → R > 0 predicted. ✅')
```
