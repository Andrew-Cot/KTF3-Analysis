### Goal
Find the physical mechanism that forces alpha_s(mu*) = b0/alpha = 0.330
from Z14 topology. Test all physically motivated ansaetze.

### Method
eta_B = e^(-N)/N from v72. Running alpha_s from one-loop QCD.
Test: which formula F(alpha_s, b0, N) = eta_B gives mu* = 0.688 GeV?

```python
import numpy as np, matplotlib.pyplot as plt
from scipy.optimize import brentq

# Known parameters
N=14.0; b0=7.0; alpha_ktf3=21.21
alpha_s_Z=0.118; alpha_em_Z=1/127.9; MZ=91.2
eta_B_obs=6.1e-8
eta_B_ktf3=np.exp(-N)/N  # = 5.94e-8
theta_R=np.pi/b0
mu_star_target=0.6880  # GeV -- what we want to derive
alpha_s_target=b0/alpha_ktf3  # = 0.3300

# One-loop running alpha_s
def alpha_s_run(mu):
    d = 1 - alpha_s_Z*b0/(2*np.pi)*np.log(MZ/mu)
    return alpha_s_Z/d if d > 0.01 else np.nan

print('='*60)
print('v78 -- Formal Derivation Scan')
print('='*60)
print(f'Target: alpha_s(mu*) = b0/alpha = {alpha_s_target:.6f}')
print(f'Target: mu* = {mu_star_target:.4f} GeV')
print(f'eta_B (v72) = e^(-N)/N = {eta_B_ktf3:.4e}')
print()
print('QUESTION: Why does Z14 topology force alpha_s(mu*) = b0/alpha?')
print()
```

```python
print('--- ANSATZ 1: Direct instanton ---')
print('In QCD, baryon number violation goes as exp(-2pi/alpha_s)')
print('If eta_B = exp(-2pi*b0/alpha_s(mu*)) / N:')
# exp(-N) = exp(-2pi*b0/alpha_s)
# alpha_s = 2pi*b0/N
alpha_s_A1 = 2*np.pi*b0/N
print(f'  alpha_s(mu*) = 2pi*b0/N = {alpha_s_A1:.4f}')
# Find mu where alpha_s = alpha_s_A1
try:
    mu_A1 = brentq(lambda mu: alpha_s_run(mu)-alpha_s_A1, 0.01, 0.5)
    print(f'  mu* = {mu_A1:.4f} GeV (target: {mu_star_target:.4f})')
    print(f'  Deviation: {abs(mu_A1-mu_star_target)/mu_star_target*100:.1f}%')
except: print('  No solution in range')
print()

print('--- ANSATZ 2: Single instanton ---')
print('If eta_B = exp(-b0/alpha_s(mu*)) / N:')
# exp(-N) = exp(-b0/alpha_s)
# alpha_s = b0/N
alpha_s_A2 = b0/N
print(f'  alpha_s(mu*) = b0/N = {b0}/{N} = {alpha_s_A2:.4f}')
try:
    mu_A2 = brentq(lambda mu: alpha_s_run(mu)-alpha_s_A2, 0.01, 1.0)
    print(f'  mu* = {mu_A2:.4f} GeV (target: {mu_star_target:.4f})')
    print(f'  Deviation: {abs(mu_A2-mu_star_target)/mu_star_target*100:.1f}%')
except: print('  No solution in range')
print()

print('--- ANSATZ 3: b0/alpha (our empirical finding) ---')
print('If alpha_s(mu*) = b0/alpha directly:')
alpha_s_A3 = b0/alpha_ktf3
print(f'  alpha_s(mu*) = b0/alpha = {b0}/{alpha_ktf3} = {alpha_s_A3:.6f}')
try:
    mu_A3 = brentq(lambda mu: alpha_s_run(mu)-alpha_s_A3, 0.1, 2.0)
    print(f'  mu* = {mu_A3:.4f} GeV (target: {mu_star_target:.4f})')
    print(f'  Deviation: {abs(mu_A3-mu_star_target)/mu_star_target*100:.2f}%')
    print(f'  This is our empirical finding -- but WHY is alpha_s = b0/alpha?')
except: print('  No solution')
print()

print('--- ANSATZ 4: Screw space winding ---')
print('Z14 screw: after N cycles, winding number = N')
print('Topological action: S = N * theta_R / alpha_s')
print('If eta_B = exp(-S) / N = exp(-N*theta_R/alpha_s) / N:')
# exp(-N) = exp(-N*theta_R/alpha_s)
# alpha_s = theta_R = pi/b0
alpha_s_A4 = np.pi/b0
print(f'  alpha_s(mu*) = theta_R = pi/b0 = {alpha_s_A4:.6f}')
try:
    mu_A4 = brentq(lambda mu: alpha_s_run(mu)-alpha_s_A4, 0.01, 0.5)
    print(f'  mu* = {mu_A4:.4f} GeV (target: {mu_star_target:.4f})')
    print(f'  Deviation: {abs(mu_A4-mu_star_target)/mu_star_target*100:.1f}%')
except: print('  No solution')
print()

print('--- ANSATZ 5: CP phase transfer ---')
print('delta_CP = theta_R/(2pi) = 1/N')
print('Sphaleron rate * delta_CP = eta_B')
print('Sphaleron rate ~ exp(-4pi/alpha_s)')
print('If eta_B = exp(-4pi/alpha_s) * (1/N):')
# exp(-N) = exp(-4pi/alpha_s)
# alpha_s = 4pi/N
alpha_s_A5 = 4*np.pi/N
print(f'  alpha_s(mu*) = 4pi/N = {alpha_s_A5:.4f}')
# This is too large -- sphaleron above TeV scale
print(f'  (Too large for confinement scale, skip)')
print()

print('--- ANSATZ 6: Z14 partition function ---')
print('Partition function of Z14 on S1: Z = sum_k exp(-k^2 * pi * alpha_s / b0)')
print('For k=1..N, dominant term: exp(-pi*alpha_s/b0)')
print('If eta_B = exp(-pi*alpha_s(mu*)/b0) / N^2:')
# exp(-N)/N = exp(-pi*alpha_s/b0)/N^2
# exp(-N)*N = exp(-pi*alpha_s/b0)
# -N + ln(N) = -pi*alpha_s/b0
# alpha_s = b0*(N-ln(N))/pi
alpha_s_A6 = b0*(N-np.log(N))/np.pi
print(f'  alpha_s(mu*) = b0*(N-ln(N))/pi = {alpha_s_A6:.4f}')
# Too large
print(f'  (Too large, skip)')
print()

print('--- ANSATZ 7: Holographic ---')
print('AdS/CFT: alpha_s ~ 1/sqrt(lambda) where lambda = g^2*N (t Hooft)')
print('If lambda = alpha * b0 = 21.21 * 7 = 148.47:')
lambda_tH = alpha_ktf3 * b0
alpha_s_A7 = 1/np.sqrt(lambda_tH)
print(f'  lambda = alpha*b0 = {lambda_tH:.2f}')
print(f'  alpha_s ~ 1/sqrt(lambda) = {alpha_s_A7:.6f}')
try:
    mu_A7 = brentq(lambda mu: alpha_s_run(mu)-alpha_s_A7, 0.1, 5.0)
    print(f'  mu* = {mu_A7:.4f} GeV (target: {mu_star_target:.4f})')
    print(f'  Deviation: {abs(mu_A7-mu_star_target)/mu_star_target*100:.1f}%')
except: print('  No solution')
print()
```

```python
print('--- SUMMARY TABLE ---')
print()
results = []
ansaetze = [
    ('2pi*b0/N (instanton)', 2*np.pi*b0/N),
    ('b0/N (single)', b0/N),
    ('b0/alpha (empirical)', b0/alpha_ktf3),
    ('pi/b0 = theta_R (screw)', np.pi/b0),
    ('1/sqrt(alpha*b0) (holographic)', 1/np.sqrt(alpha_ktf3*b0)),
    ('(N-ln(N))/(2*pi*b0) (Euler)', (N-np.log(N))/(2*np.pi*b0)),
]
print(f'{"Ansatz":<35} {"alpha_s":>8} {"mu* [GeV]":>10} {"Dev %":>8}')
print('-'*65)
for name, alpha_s_pred in ansaetze:
    try:
        mu = brentq(lambda mu: alpha_s_run(mu)-alpha_s_pred, 0.01, 5.0)
        dev = abs(mu-mu_star_target)/mu_star_target*100
        marker = ' <-- BEST' if dev < 5 else ''
        print(f'{name:<35} {alpha_s_pred:>8.4f} {mu:>10.4f} {dev:>8.1f}%{marker}')
        results.append((name, alpha_s_pred, mu, dev))
    except:
        print(f'{name:<35} {alpha_s_pred:>8.4f} {"out of range":>10}')
print()
print(f'Target mu* = {mu_star_target:.4f} GeV')
print()

# Find best
if results:
    best = min(results, key=lambda x: x[3])
    print(f'BEST ANSATZ: {best[0]}')
    print(f'  alpha_s = {best[1]:.6f}')
    print(f'  mu* = {best[2]:.4f} GeV ({best[3]:.1f}% deviation)')
print()

# Physical interpretation of best
print('--- PHYSICAL INTERPRETATION OF BEST ---')
print()
print('If alpha_s(mu*) = b0/alpha:')
print(f'  b0 = QCD beta coefficient = {b0}')
print(f'  alpha = ln(1/eta_B) = {alpha_ktf3}')
print(f'  This says: at mu*, the QCD coupling equals')
print(f'  the ratio of QCD color structure (b0)')
print(f'  to cosmological information content (alpha)')
print()
print('  Rewrite: alpha_s(mu*) x alpha = b0')
print(f'  = {alpha_s_target:.4f} x {alpha_ktf3:.2f} = {alpha_s_target*alpha_ktf3:.4f} ~ b0 = {b0}')
print()
print('MEANING:')
print('  At the confinement scale mu*, the product')
print('  alpha_s x alpha = b0 = 7')
print('  This is a CONSERVATION LAW of the Z14 geometry:')
print('  strong coupling x cosmological entropy = QCD color charge')
```
