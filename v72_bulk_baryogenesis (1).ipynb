## 1. Known Parameters (no free parameters)

````python
# --- QCD parameters ---
N_f = 6                          # number of quark flavours
b0 = 11 - 2*N_f/3               # QCD beta function = 7
N = 2 * b0                      # Z14 cycle number = 14

# --- Geometric parameters (from CMB) ---
T1_Mpc = 1660.0                  # fundamental period [Mpc]
theta_R = np.pi / b0             # rotation per cycle [rad] = 25.714 deg
theta_R_deg = np.degrees(theta_R)

# --- Electroweak ---
alpha_w = 1/30                   # weak coupling

# --- Observed ---
eta_B_observed = 6.1e-8          # baryon asymmetry (Planck 2018)
alpha_observed = np.log(1/eta_B_observed)  # = 21.21

print(f'QCD beta function b0     = {b0}')
print(f'Cycle number N = 2*b0    = {N}')
print(f'θ_R = π/b0               = {theta_R_deg:.4f} deg')
print(f'T1                       = {T1_Mpc} Mpc')
print(f'Renewal period N*T1      = {N*T1_Mpc:.0f} Mpc')
print(f'η_B observed             = {eta_B_observed:.2e}')
print(f'α = ln(1/η_B)            = {alpha_observed:.4f}')
````

## 2. Mechanism A: Bulk Leakage

Two KTF³ separated by bulk distance d = N × T₁ = 23,240 Mpc  
Bulk coupling: g_bulk = e^(-d/T₁) = e^(-N)  
Asymmetry from rotation: (1 - cos(θ_R))  

```
[KTF³ #1] ←— d = N×T₁ —→ [KTF³ #2]
    our universe              rotated by θ_R
```

````python
# Bulk coupling
g_bulk = np.exp(-N)              # e^(-14)

# Rotational asymmetry between the two KTF3
asymmetry = 1 - np.cos(theta_R)  # (1 - cos(π/b0))

# Predicted eta_B from Mechanism A
eta_B_A = g_bulk * asymmetry

print('--- MECHANISM A: Bulk Leakage ---')
print(f'Bulk distance d          = N×T₁ = {N*T1_Mpc:.0f} Mpc')
print(f'Bulk coupling g_bulk     = e^(-{N}) = {g_bulk:.3e}')
print(f'Rotational asymmetry     = 1 - cos(π/b0) = {asymmetry:.6f}')
print(f'η_B predicted (A)        = {eta_B_A:.3e}')
print(f'η_B observed             = {eta_B_observed:.3e}')
print(f'Ratio predicted/observed = {eta_B_A/eta_B_observed:.3f}')
print(f'Deviation                = {abs(eta_B_A - eta_B_observed)/eta_B_observed * 100:.1f}%')
````

## 3. Mechanism B: Cycle Transfer

CP-asymmetry does not thermalize — it transfers to the next Z14 cycle.  
After 14 cycles: spinor phase = -1 (v68 confirmed).  
The leak per cycle = 1/N (equal distribution over 14 cycles).  
Suppression by weak coupling α_w per transfer.

```
Cycle 1 → Cycle 2 → ... → Cycle 14 → back to Cycle 1
  each transfer leaks (1/N) × α_w of CP-asymmetry
```

````python
# Cycle transfer mechanism
# Each cycle transfers fraction (1/N) with weak suppression
# After N cycles with spinor phase -1: net = epsilon_CP × (alpha_w/N)^N

epsilon_CP = np.sin(theta_R / 2)  # CP asymmetry from topology
transfer_per_cycle = alpha_w / N  # suppression per cycle

# After N cycles
eta_B_B = epsilon_CP * (transfer_per_cycle)**2  # two-body suppression

# Alternative: spinor phase -1 after 14 cycles means sign flip
# Net surviving fraction after round trip:
f_cycle = (1 - np.exp(-N * alpha_w)) / N
eta_B_B2 = epsilon_CP * alpha_w**2 * f_cycle

print('--- MECHANISM B: Cycle Transfer ---')
print(f'ε_CP = sin(θ_R/2)        = {epsilon_CP:.6f}')
print(f'Transfer per cycle       = α_w/N = {transfer_per_cycle:.6f}')
print(f'η_B predicted (B1)       = {eta_B_B:.3e}')
print(f'η_B predicted (B2)       = {eta_B_B2:.3e}')
print(f'η_B observed             = {eta_B_observed:.3e}')
````

## 4. Combined: Bulk + Cycle

Both mechanisms act simultaneously.  
Bulk leakage sets the overall scale.  
Cycle transfer provides the phase coherence.

````python
# Combined mechanism
# Bulk gives the suppression scale
# Cycle gives the phase factor
# Spinor phase after 14 cycles = -1 (v68) adds factor of 2 (interference)

spinor_phase = -0.999998  # from v68
interference = abs(1 - spinor_phase) / 2  # = ~1, constructive

eta_B_combined = g_bulk * asymmetry * (1 + abs(spinor_phase)) / 2

print('--- COMBINED: Bulk + Cycle ---')
print(f'Spinor phase after 14 cycles = {spinor_phase:.6f} (v68)')
print(f'Interference factor          = {interference:.6f}')
print(f'η_B combined                 = {eta_B_combined:.3e}')
print(f'η_B observed                 = {eta_B_observed:.3e}')
print(f'Ratio                        = {eta_B_combined/eta_B_observed:.3f}')
````
