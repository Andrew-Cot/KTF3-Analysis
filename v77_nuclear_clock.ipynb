```python
import numpy as np, matplotlib.pyplot as plt
N=14.0; b0=7.0; alpha_ktf3=21.21; T1_Mpc=1660.0
alpha_s_Z=0.118; alpha_em_Z=1/127.9; MZ=91.2
eta_B=6.1e-8; theta_R=np.pi/b0
ratio_ktf3 = b0/alpha_ktf3
log_ratio = (alpha_s_Z/ratio_ktf3-1)*2*np.pi/(alpha_s_Z*b0)
mu_star = MZ*np.exp(log_ratio)
T_renewal_yr = N*T1_Mpc*3.086e22/3e8/(365.25*24*3600)
drift_ktf3 = 1/T_renewal_yr
ratio_ktf3_pred = alpha_em_Z*alpha_ktf3/b0
print(f'b0={b0}, N={N}, alpha={alpha_ktf3}')
print(f'theta_R = pi/b0 = {np.degrees(theta_R):.4f} deg')
print()
print('PREDICTION 1: Alpha drift')
print(f'  T_renewal = N*T1 = {N*T1_Mpc:.0f} Mpc = {T_renewal_yr:.2e} years')
print(f'  KTF3-R drift bound: {drift_ktf3:.2e} per year')
print(f'  Peik threshold:     1e-18 per year')
print(f'  --> KTF3-R predicts NO alpha drift (10^{int(np.log10(1e-18/drift_ktf3))} below threshold)')
print()
print('PREDICTION 2: New physics scale mu*')
print(f'  alpha_s(mu*) = b0/alpha = {ratio_ktf3:.6f}')
print(f'  mu* = {mu_star:.0f} GeV = {mu_star/1000:.1f} TeV')
print(f'  Compare: k_AdS = 32 TeV (Radojicic 2026)')
print(f'  mu*/k_AdS = {mu_star/1000/32:.3f}')
print()
print('PREDICTION 3: alpha_em/alpha_s ratio fixed by geometry')
print(f'  alpha_em(MZ) x alpha / b0 = {ratio_ktf3_pred:.6f}')
print()
print('HONEST ASSESSMENT:')
print('  P1 (no drift): clean but weak -- null hard to falsify')
print('  P2 (mu* scale): interesting -- not formally derived')
print('  P3 (ratio): speculative -- needs formal derivation')
print()
# Plot
fig, axes = plt.subplots(1, 2, figsize=(14,6))
fig.suptitle('v77 -- KTF3-R Predictions for Nuclear Clock (Peik/PTB)', fontsize=13, fontweight='bold')
ax1=axes[0]
mu_range=np.logspace(-1,5,500)
alpha_s_run=[alpha_s_Z/(1+alpha_s_Z*b0/(2*np.pi)*np.log(mu/MZ)) if alpha_s_Z/(1+alpha_s_Z*b0/(2*np.pi)*np.log(mu/MZ))>0 else np.nan for mu in mu_range]
ax1.semilogx(mu_range,alpha_s_run,'b-',lw=2,label='alpha_s running (1-loop)')
ax1.axhline(ratio_ktf3,color='red',linestyle='--',lw=2,label=f'KTF3-R: b0/alpha={ratio_ktf3:.4f}')
ax1.axvline(mu_star,color='green',linestyle=':',lw=2,label=f'mu*={mu_star:.0f} GeV')
ax1.scatter([mu_star],[ratio_ktf3],color='green',s=150,zorder=5)
ax1.set_xlabel('Energy [GeV]'); ax1.set_ylabel('alpha_s(mu)')
ax1.set_title('Running alpha_s and KTF3-R scale mu*')
ax1.legend(fontsize=9); ax1.grid(True,alpha=0.3)
ax1.set_xlim(0.1,1e5); ax1.set_ylim(0,0.5)
ax2=axes[1]
labs=['Peik threshold','Standard cosmology','KTF3-R']
vals=[1e-18,1e-21,drift_ktf3]
cols=['orange','blue','green']
bars=ax2.bar(labs,vals,color=cols,alpha=0.8,edgecolor='black')
ax2.set_yscale('log'); ax2.set_ylabel('|delta_alpha/alpha| per year')
ax2.set_title('Alpha drift: KTF3-R vs Peik threshold')
ax2.grid(True,alpha=0.3,axis='y')
for bar,val in zip(bars,vals): ax2.text(bar.get_x()+bar.get_width()/2,val*3,f'{val:.0e}',ha='center',fontsize=10,fontweight='bold')
ax2.text(0.5,0.95,'KTF3-R: NO drift predicted',transform=ax2.transAxes,ha='center',va='top',color='green',fontweight='bold')
plt.tight_layout()
plt.savefig('v77_nuclear_clock.png',dpi=150,bbox_inches='tight')
plt.show()
print('GitHub: github.com/Andrew-Cot/KTF3-Analysis | OSF: osf.io/42nks')
```
